```python
import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
(X_train,y_train),(X_test,y_test)=keras.datasets.fashion_mnist.load_data()
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

```python
X_train[0]
```

```python
y_train[0]
```

```python
plt.imshow(X_train[0],cmap='Greys')
```

```python
class_labels=["T-shirt/top","Trouser","Pullover","Dress","Coat","Sandal","Shirt","Sneaker","Bag","Ankle boot"]
```

```python
class_labels
```

```python
plt.figure(figsize=(16,16))
for i in range(25):
  plt.subplot(5,5,i+1)
  plt.imshow(X_train[i],cmap='Greys')
  plt.axis('off')
  plt.title(class_labels[y_train[i]]+'='+str(y_train[i]),fontsize=20)
```

```python
X_train=X_train/255
X_test=X_test/255
```

```python
X_train[0]
```

```python
model=keras.models.Sequential([
                        keras.layers.Flatten(input_shape=[28,28]),
                        keras.layers.Dense(units=32,activation='relu'),
                        keras.layers.Dense(units=10,activation='softmax')   
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=10)
```

```python
model.evaluate(X_test,y_test)
```

```python
 y_predict=model.predict(X_test)
```

```python
y_predict[0].round(2)
```

```python
np.argmax(y_predict[0].round(2))
```

```python
y_test[0]
```

```python
plt.figure(figsize=(16,16))
for i in range(25):
  plt.subplot(5,5,i+1)
  plt.imshow(X_test[i],cmap="Greys")
  plt.axis('off')
  plt.title("Actual={}\n predicted={}".format(class_labels[y_test[i]],class_labels[np.argmax(y_predict[i])]))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,[np.argmax(i) for i in y_predict])
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(cm,annot=True,fmt="d")
```

```python
from sklearn.metrics import classification_report
cr=classification_report(y_test,[np.argmax(i) for i in y_predict],target_names=class_labels,)
```

```python
print(cr)
```
